```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_kway(title, maximum, mean):
    plt.plot(maximum)
    plt.plot(mean)
    plt.ylim(0,100)
    plt.ylabel("Fitness")
    plt.xlabel("Generations")
    plt.legend(["Maxmimum", "Mean"])
    plt.title(title + str(round(np.max(maximum), 2)))
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().yaxis.set_ticks_position('left')
    plt.gca().xaxis.set_ticks_position('bottom')
    plt.show()
```
